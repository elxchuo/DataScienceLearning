# Introducción a los Árboles de Decisión

Los **árboles de decisión** son uno de los algoritmos más populares y versátiles en el campo del aprendizaje automático (*machine learning*). Se utilizan tanto para tareas de predicción como para la toma de decisiones, gracias a su estructura intuitiva que imita el razonamiento humano. Un árbol de decisión está compuesto por:

- **Nodo raíz**: Punto de partida.
- **Nodos internos**: Puntos de decisión basados en características del conjunto de datos.
- **Hojas**: Resultados o predicciones finales.

### Ventajas
- Fáciles de visualizar y entender, incluso sin conocimientos profundos de matemáticas o programación.
- Pueden manejar datos numéricos y categóricos.
- Robustos ante valores faltantes o ruido.

### Desventajas
- Tendencia al sobreajuste (*overfitting*) si no se controlan.
- Sensibles a pequeños cambios en los datos (inestabilidad).

Esto ha llevado al desarrollo de variantes avanzadas como *Random Forests* o *Gradient Boosting*.

---

# Clasificación de los Árboles de Decisión

## 1. Árboles de Regresión
Se usan cuando la variable objetivo es **continua** (valores numéricos).

- **Cómo funcionan**:
  - Divide el espacio de características en regiones basadas en umbrales.
  - En cada hoja, el valor predicho es el promedio (o mediana) de la variable objetivo en esa región.
  - Criterio de división: Minimizar la **varianza** o el **error cuadrático medio (MSE)**.

- **Ejemplo**:
  - Problema: Predecir el consumo de energía de un edificio.
  - Variables: Tamaño, ocupantes, temperatura exterior.
  - Resultado: Valor en kWh.

- **Algoritmo representativo**: CART (*Classification and Regression Trees*) en modo regresión.

## 2. Árboles de Clasificación
Se usan cuando la variable objetivo es **categórica** (etiquetas o clases).

- **Cómo funcionan**:
  - Divide el espacio de características en regiones basadas en reglas.
  - En cada hoja, la predicción es la clase más frecuente (o probable).
  - Criterios de división: **Impureza de Gini**, **entropía** (ganancia de información), o **error de clasificación**.

- **Ejemplo**:
  - Problema: Clasificar si un cliente comprará un producto.
  - Variables: Edad, ingresos, historial de compras.
  - Resultado: "Sí" o "No".

- **Algoritmos representativos**: CART en modo clasificación, ID3, C4.5.

---

# Otros Grupos y Clasificaciones

## 3. Árboles Básicos vs. Ensamblados
- **Árboles Básicos**: Árboles individuales (CART, ID3, C4.5). Simples, pero propensos al sobreajuste.
- **Árboles Ensamblados**: Combinan múltiples árboles para mayor precisión y estabilidad.
  - **Random Forest**: Usa subconjuntos aleatorios de datos y características, promediando (regresión) o votando (clasificación).
  - **Gradient Boosting**: Construye árboles secuencialmente corrigiendo errores (ej: XGBoost, LightGBM).

## 4. Según el Criterio de División
- **Impureza de Gini**: Usada en CART, mide la probabilidad de error en clasificación.
- **Entropía/Ganancia de Información**: Usada en ID3 y C4.5, mide el desorden.
- **Reducción de Varianza**: Usada en regresión para minimizar la dispersión.

## 5. Según la Complejidad
- **Árboles Podados**: Limitados para evitar sobreajuste (poda pre o post-entrenamiento).
- **Árboles Sin Podar**: Divisiones hasta que las hojas sean puras, pero suelen sobreajustarse.

---

# Notas Adicionales
- **Ventajas**: Interpretabilidad, no necesitan normalización, capturan relaciones no lineales.
- **Desventajas**: Inestables sin ensamblado, propensos al sobreajuste.
- **Aplicaciones**: Diagnóstico médico, análisis financiero, segmentación de clientes.

El proceso para construir y usar un modelo tiene **cuatro pasos**:

1. **Definir**:

   - Revisas qué tan acertadas son las predicciones del modelo. ¿Es bueno o necesita mejorar?

# ¿Qué es el MSE inicial?

El **MSE inicial** es el error cuadrático medio de todos los datos antes de hacer cualquier división en el árbol de decisión. Es decir, se calcula asumiendo que, sin dividir, la predicción para todos los puntos es el **promedio de todos los valores de `y`** (el *target*, en tu caso, los precios de las casas). Este valor es el punto de partida que el árbol intentará mejorar con las divisiones.

El MSE se calcula como:

$$MSE = \frac{1}{n} \sum (valor_{real} - valor_{predicho})^2$$

### Ejemplo basico 
Ocupar el promedio de los valores de las casas para saber un aproximado cuanto valor sera.

Promedio = (200,000 + 300,000 + 500,000 + 700,000 + 900,000) / 5
         = 2,600,000 / 5
         = 520,000

Sin dividir, el modelo predice 520,000 para todas las casas.

Paso 2: Calcular las diferencias al cuadrado
Restamos el promedio (520,000) a cada valor real y elevamos al cuadrado:

200,000 - 520,000 = -320,000 → (-320,000)² = 102,400,000,000
300,000 - 520,000 = -220,000 → (-220,000)² = 48,400,000,000
500,000 - 520,000 = -20,000 → (-20,000)² = 400,000,000
700,000 - 520,000 = 180,000 → (180,000)² = 32,400,000,000
900,000 - 520,000 = 380,000 → (380,000)² = 144,400,000,000
Paso 3: Sumar las diferencias al cuadrado

El MSE inicial es 65,600,000,000. Esto mide qué tan dispersos están los precios respecto al promedio sin ninguna división.

# El Problema con las Puntuaciones "Dentro de la Muestra" (In-Sample Scores)

Cuando usamos los **mismos datos** para entrenar un modelo y luego evaluarlo, obtenemos lo que se llama una **puntuación "dentro de la muestra"** (*in-sample score*). Esto significa que tomamos un solo grupo de datos (una "muestra") para construir el modelo y medir qué tan bueno es. Pero esto tiene un problema importante. Vamos a entenderlo con un ejemplo sencillo.

---

## ¿Por qué es malo usar los mismos datos?

### Ejemplo para entenderlo
Imagina que estás trabajando con datos de casas en un mercado inmobiliario enorme. En la realidad, el **color de la puerta** no tiene nada que ver con el precio de una casa (puerta verde, azul o roja, da igual).

Pero supongamos que, por pura casualidad, en el pequeño grupo de datos que usaste para entrenar tu modelo (digamos, 100 casas), **todas las casas con puertas verdes eran muy caras**. El modelo va a buscar patrones para predecir precios y "aprenderá" que "puertas verdes = casas caras". Entonces, siempre que vea una casa con puerta verde, dirá que es cara.

#### ¿Qué pasa al evaluar?
Si pruebas el modelo con esos mismos 100 datos, va a parecer que es súper bueno, porque predice bien los precios de esas casas. Pensarás: "¡Mi modelo es increíble!".

#### El problema con datos nuevos
Ahora, si usas el modelo con **datos nuevos** (casas que no estaban en esos 100 originales), y en el mundo real el color de la puerta no importa, el modelo va a fallar. Por ejemplo, si una casa con puerta verde es barata, el modelo dirá que es cara y se equivocará. Esto muestra que el modelo no es tan bueno como parecía.

---

## ¿Por qué importa esto?
Los modelos de *machine learning* son útiles porque queremos que hagan **predicciones buenas en datos nuevos**, no solo en los datos que ya tenemos. Si medimos el rendimiento del modelo con los mismos datos que usamos para entrenarlo, no sabemos cómo va a funcionar en la práctica. Es como hacer trampa en un examen: si ya sabes las respuestas porque las estudiaste, no significa que entiendas el tema.

---

## La Solución: Datos de Validación

Para evitar este problema, **separamos los datos** en dos partes:

1. **Datos de entrenamiento**: Los usamos para construir el modelo.
2. **Datos de validación**: Los guardamos y los usamos después para probar el modelo. Estos datos son "nuevos" para el modelo porque no los vio durante el entrenamiento.

Así, podemos medir qué tan bien funciona el modelo con datos que no conoce, lo que nos da una idea realista de su rendimiento en el mundo real.

### Ejemplo sencillo
- Tienes datos de 100 casas.
- Usas 80 para entrenar el modelo (datos de entrenamiento).
- Guardas 20 para probarlo después (datos de validación).
- Si el modelo predice bien los precios de esas 20 casas que no vio antes, puedes confiar en él.

---

## Resumen en Palabras Simples
- Usar los mismos datos para entrenar y evaluar engaña: parece que el modelo es perfecto, pero no sabes si sirve con datos nuevos.
- Para hacerlo bien, separa los datos: unos para entrenar y otros para probar.
- Los **datos de validación** te dicen la verdad sobre cómo funciona tu modelo en la práctica.

---

## Model validation

### Solución
La forma más fácil de hacer esto es:
- **Separar algunos datos** desde el principio y no usarlos para entrenar.
- Usar esos datos guardados para probar el modelo y ver qué tan bien predice cosas que no ha visto antes.

---
La biblioteca **`scikit-learn`** tiene una función llamada **`train_test_split`** que divide los datos en dos partes:
- **Datos de entrenamiento**: Los usas para enseñar al modelo.
- **Datos de validación**: Los usas para probar el modelo y medir su error.

Con los datos de validación, puedes calcular el **error absoluto medio** (*mean_absolute_error*), que te dice qué tan lejos están las predicciones de los valores reales.

```python
from sklearn.model_selection import train_test_split

# split data into training and validation data, for both features and target
# The split is based on a random number generator. Supplying a numeric value to
# the random_state argument guarantees we get the same split every time we

# run this script.
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)
# Define model
melbourne_model = DecisionTreeRegressor()
# Fit model
melbourne_model.fit(train_X, train_y)

# get predicted prices on validation data
val_predictions = melbourne_model.predict(val_X)
print(mean_absolute_error(val_y, val_predictions))
```

## Ejemplo Kaggle Iowa completo de regresor decision tree

```python
# Paquetes 
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split

path = r'C:/Users/JoseCortes/OneDrive - Tellus - Mining/Escritorio/Ciencia de Datos/DataScienceLearning/data/Kaggle/home-data-for-ml-course/train.csv'
home_data = pd.read_csv(path)

# Objeto predictivo (Precio de Venta)
y = home_data['SalePrice']

# X Caracteristicas del modelo (Features)
features = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
X = home_data[features]

# Vdividirlo entre datos de validacion y datos de entramiento. !!
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1)

# Creacion del modelo regressor arbol tree
iowa_model = DecisionTreeRegressor(random_state=1)

# Entraniento del modelo. (Con las variables creadas en validation data.)
iowa_model.fit(train_X, train_y)

# Medir las validaciones de tu modelo !
val_predicciones = iowa_model.predict(val_X)

# Medir el MAE (Mean Absolute Error) de las validaciones 
val_mae = mean_absolute_error(val_predicciones, val_y)
print("Validation MAE when not specifying max_leaf_nodes: {:,.0f}".format(val_mae))

'''' Usaremos otros elementos como el max_leaf_nodes = 100 para apreciar que sucede con la validacion del modelo'''
iowa_model_max_leaf = DecisionTreeRegressor(max_leaf_nodes=100, random_state=1)
# Entrar el modelo con los datos de validacion y entrenamiento
iowa_model_max_leaf.fit(train_X, train_y)
# crear las predicciones del modelo con los datos de validacion.
val_predicciones_max_leaf = iowa_model_max_leaf.predict(val_X)
# Revisar el MSE  de la validacion.
val_mae_max_leaf = mean_absolute_error(val_predicciones_max_leaf, val_y)
print("Validation MAE for best value of max_leaf_nodes: {:,.0f}".format(val_mae_max_leaf))
```

## Reemplazemos el Deicsion Tree por un Random Forest.



```python
from sklearn.ensemble import RandomForestRegressor

# Definicion del modelo
rf_model = RandomForestRegressor(random_state=1)

# entrenar el model de random forest Fit you model
rf_model.fit(train_X, train_y)

#Predicciones del modelo random forest
rf_val_predictions = rf_model.predict(val_X)

# Calculo de MAE para validacion del modelo.
rf_val_mae = mean_absolute_error(rf_val_predictions, val_y)

print("Validation MAE for Random Forest Model: {}".format(rf_val_mae))
```

```python
from sklearn.ensemble import RandomForestRegressor

# Define the models
model_1 = RandomForestRegressor(n_estimators=50, random_state=0)
model_2 = RandomForestRegressor(n_estimators=100, random_state=0)
model_3 = RandomForestRegressor(n_estimators=100, criterion='absolute_error', random_state=0)
model_4 = RandomForestRegressor(n_estimators=200, min_samples_split=20, random_state=0)
model_5 = RandomForestRegressor(n_estimators=100, max_depth=7, random_state=0)

models = [model_1, model_2, model_3, model_4, model_5]
```

To select the best model out of the five, we define a function score_model() below. This function returns the mean absolute error (MAE) from the validation set. Recall that the best model will obtain the lowest MAE. (To review mean absolute error, look here.)

Run the code cell without changes.

```python
from sklearn.metrics import mean_absolute_error

# Function for comparing different models
def score_model(model, X_t=X_train, X_v=X_valid, y_t=y_train, y_v=y_valid):
    model.fit(X_t, y_t)
    preds = model.predict(X_v)
    return mean_absolute_error(y_v, preds)

for i in range(0, len(models)):
    mae = score_model(models[i])
    print("Model %d MAE: %d" % (i+1, mae))
```
